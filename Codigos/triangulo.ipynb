```python
#1) Centroide desde vértices
def centroide_desde_vertices(A, B, C):
    A, B, C = V(A), V(B), V(C)
    G = (A + B + C) / 3
    return G
```

```python
#2) Centroide desde lados AB y BC, con A conocido
def centroide_desde_lados(AB, BC, A=None):
    AB, BC = V(AB), V(BC)

    A = V(A) if A is not None else Matrix.zeros(AB.rows, 1)
    B = A + AB
    C = B + BC
    G = (A + B + C) / 3
    return G
```
